```python
#data cleaning for species column - removing rows, where sharks' involment was not prooved

values_to_remove = [
    'doubtful / unconfirmed attack / unable to verify in local records',
    'injuries not caused by a shark',
    'no shark involvement',
    'no shark invovlement - it ws a publicity stunt',
    'no shark invovlement ',
    'questionable incident, said to involve a 6\' shark',
    'questionable incident; reported as shark attack but thought to involve a pinniped instead ',
    'reported as a shark attack, the story was a hoax',
    'said to involve an 8\' shark but more likely damage caused by debris',
    'Shark involment not confirmed',
    'shark involvement  questionable',
    'shark involvement highly doubtful',
    'shark involvement not confirmed & highly unlikely',
    'shark involvement not confirmed',
    'shark involvement not confirmes',
    'shark involvement  not confirmed',
    'shark involvement prior to death not confirmed',
    'shark involvement prior to death still to be determined',
    'shark involvement prior to death suspected but not confirmed',
    'shark involvement prior to death unconfired',
    'Shark involvement prior to death unconfirmed',
    'Shark involvement prior to death was not confirmed',
    'shark involvement prior to deaths unconfirmed',
    'shark involvement probable',
    'shark involvement questionable',
    'shark involvement prior to death could not be determined',
    'shark involvement suspected but not confirmed',
    'shark involvement prior to death remains unconfirmed',
    'shark involvement unconfirmed but considered probable',
    'thought to involve a 3\' to 4\' shark, but shark involvement not confirmed',

]

shark_attack_new = shark_attack[~shark_attack['species'].isin(values_to_remove)]
shark_attack_new['species'].nunique()
```

```python
shark_attack_new['species'].value_counts()
```

```python
#data cleaning for species column - replacing species, which containes similar names

shark_attack_new = replace_species(shark_attack_new)
shark_attack_new['species'].nunique()
```

```python
#data cleaning for species column - here we delated the remained values, that contains numbers only

import numpy as np

shark_attack_new['species'] = shark_attack_new['species'].where(~shark_attack_new['species'].str.contains('\d', na=False), np.nan)
shark_attack_new['species'].nunique()
```

```python
#data cleaning for species column - replacing remined invalid data, such as ' ' with NaN.

words_to_replace = ['Barracuda (not shark)', 'small', 'school', 'large', 'pack', 'unknown', 'nan', 'does not',
                    ' ', '  ', 'unidentified', '\xa0', 'questionable', 'invalid', ]

shark_attack_modified = replace_values_with_nan(shark_attack_new, words_to_replace)

shark_attack_modified['species'].nunique()
```
